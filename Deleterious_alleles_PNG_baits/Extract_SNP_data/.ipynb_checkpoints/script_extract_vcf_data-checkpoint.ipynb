```python
# Make a simple script to extract this fields (without separating them at first).
import pandas as pd
import gzip

def read_vcf_to_dataframe(vcf_file):
    opener = gzip.open if vcf_file.endswith(".gz") else open

    with opener(vcf_file, "rt") as f:
        for line in f:
            if line.startswith("#CHROM"):
                header = line.strip().lstrip("#").split("\t")
                break

        rows = []
        for line in f:
            if line.startswith("#"):
                continue
            rows.append(line.strip().split("\t"))

    df = pd.DataFrame(rows, columns=header)

    # Basic type conversion
    df["POS"] = df["POS"].astype(int)
    df["QUAL"] = pd.to_numeric(df["QUAL"], errors="coerce")

    # ------------------
    # Parse INFO field
    # ------------------
    info_dicts = []
    for info in df["INFO"]:
        d = {}
        for item in info.split(";"):
            if "=" in item:
                k, v = item.split("=", 1)
                d[k] = v
            else:
                d[item] = True
        info_dicts.append(d)

    info_df = pd.DataFrame(info_dicts)

    # ------------------
    # Parse samples (keep one column per sample)
    # ------------------
    sample_cols = header[9:]

    def parse_sample(sample_str, keys):
        values = sample_str.split(":")
        return dict(zip(keys, values))

    format_keys = df["FORMAT"].iloc[0].split(":")

    samples_df = pd.DataFrame({
        sample: df[sample].apply(lambda x: parse_sample(x, format_keys))
        for sample in sample_cols
    })

    # ------------------
    # Final dataframe
    # ------------------
    fixed_cols = ["CHROM", "POS", "ID", "REF", "ALT", "QUAL", "FILTER"]

    final_df = pd.concat(
        [
            df[fixed_cols],
            info_df,
            samples_df
        ],
        axis=1
    )

    return final_df

```

## 2. Extract the genotypes to make a table

```python
# Fields to keep?
# https://gatk.broadinstitute.org/hc/en-us/articles/360035531692-VCF-Variant-Call-Format
# https://gatk.broadinstitute.org/hc/en-us/articles/360035531672-Haplotype-map-format?utm_source=chatgpt.com 

# GT:AD:DP:GQ:PGT:PID:PL:PS
##FORMAT=<ID=GT,Number=1,Type=String,Description="Genotype">
##FORMAT=<ID=AD,Number=R,Type=Integer,Description="Allelic depths for the ref and alt alleles in the order listed">
##FORMAT=<ID=DP,Number=1,Type=Integer,Description="Approximate read depth (reads with MQ=255 or with bad mates are filtered)">
##FORMAT=<ID=GQ,Number=1,Type=Integer,Description="Genotype Quality">
##FORMAT=<ID=PGT,Number=1,Type=String,Description="Physical phasing haplotype information, describing how the alternate alleles are phased in relation to one another">
##FORMAT=<ID=PID,Number=1,Type=String,Description="Physical phasing ID information, where each unique ID within a given sample (but not across samples) connects records within a phasing group">
##FORMAT=<ID=PL,Number=G,Type=Integer,Description="Normalized, Phred-scaled likelihoods for genotypes as defined in the VCF specification">
## PS= phase set

# Example:
# 0/0:1,0:1:3:0,3,12
```

```python
# Make the table showing genotypes of each samples for each conserved gene.

def extract_genotypes_with_pos(df):
    sample_cols = [c for c in df.columns if isinstance(df[c].iloc[0], dict)]

    gt_df = pd.DataFrame({
        sample: df[sample].apply(lambda x: x.get("GT", None))
        for sample in sample_cols
    })

    return pd.concat([df[["CHROM", "POS"]], gt_df], axis=1)
```

## 3. On another dataframe, extract the SIFT annotations.

```python
# 1. Define ANN
ann_columns = [
    "Allele",
    "Annotation",
    "Impact",
    "Gene_Name",
    "Gene_ID",
    "Feature_Type",
    "Feature_ID",
    "Transcript_BioType",
    "Rank",
    "HGVS_c",
    "HGVS_p",
    "cDNA_pos",
    "CDS_pos",
    "AA_pos",
    "Distance",
    "Errors"
]

# Define the function
#
def parse_ann_field(ann_string):
    if pd.isna(ann_string):
        return pd.Series([None] * len(ann_columns))
    
    # Take first annotation if multiple
    first_ann = ann_string.split(",")[0]
    
    values = first_ann.split("|")
    
    # Pad if missing fields
    if len(values) < len(ann_columns):
        values += [None] * (len(ann_columns) - len(values))
    
    return pd.Series(values[:len(ann_columns)])

```

```python
ann_df = df["ANN"].apply(parse_ann_field)
ann_df.columns = ann_columns
ann_df
```

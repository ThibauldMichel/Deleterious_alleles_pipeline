# Group BLAST results so that “nearly identical” hits (same qseqid and sseqid, but with close values for send and slen) are clustered, and only the best hit per cluster (highest pident and bitscore) is kept.
  

```python
import pandas as pd
import numpy as np

# Parameters
tol_send = 10000000
tol_slen = 10000000


# Sort for consistent cluster numbering
df = df.sort_values(["qseqid", "sseqid", "send", "slen"]).reset_index(drop=True)

# Create cluster_id column
cluster_ids = np.zeros(len(df), dtype=int)
cluster_counter = 0

# Loop over each (qseqid, sseqid) group
for _, group_idx in df.groupby(["qseqid", "sseqid"]).groups.items():
    group = df.loc[group_idx]
    assigned = np.zeros(len(group), dtype=bool)
    
    for i, row in group.iterrows():
        if not assigned[group.index.get_loc(i)]:
            cluster_counter += 1
            cluster_ids[i] = cluster_counter
            ref_send, ref_slen = row["send"], row["slen"]
            mask = (
                (~assigned) &
                (np.abs(group["send"] - ref_send) <= tol_send) &
                (np.abs(group["slen"] - ref_slen) <= tol_slen)
            )
            assigned[mask.values] = True
            cluster_ids[group.index[mask]] = cluster_counter

# Add column to DataFrame
df["cluster_id"] = cluster_ids

# Save to CSV
df.to_csv("results_blast_clustered.csv", index=False)
```

```python
df
```

```python
# Select best hit per cluster (highest pident, then highest bitscore)
best_hits = (
    df.sort_values(["pident", "bitscore"], ascending=[False, False])
      .groupby("cluster_id", as_index=False)
      .first()
)

# Save to CSV
best_hits.to_csv("results_blast_best_per_cluster.csv", index=False)
```

```python
# This works regardless of which is larger (forward or reverse strand)
best_hits['sequence_length'] = abs(best_hits['send'] - best_hits['sstart']) + 1

# Or if you want to mark strand direction:
best_hits['strand'] = np.where(best_hits['sstart'] <= best_hits['send'], '+', '-')
```

```python
# Trying to isolate duplicates to check where they are
# This keeps ALL duplicate rows (not just the first or last occurrence)
duplicated_rows = best_hits[best_hits["qseqid"].duplicated(keep=False)]
duplicated_rows
```

```python
# Filter out the duplicated matches
single_hits = best_hits[~best_hits["qseqid"].duplicated(keep=False)]
# Save to CSV
single_hits.to_csv("results_blast_best_per_cluster_Only_one_match.csv", index=False)
```
